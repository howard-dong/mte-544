```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plot_path(df, title):
    plt.plot(df['x'].to_numpy(), df['y'].to_numpy())
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
```

```python
plot_path(pd.read_csv("odom_content_line.csv", delimiter=', '), "Straight Line")
```

```python
plot_path(pd.read_csv("odom_content_spiral.csv", delimiter=', '), "Spiral")
```

```python
plot_path(pd.read_csv("odom_content_circle.csv", delimiter=', '), "Circle")
```
